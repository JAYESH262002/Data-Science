```python
for feature in numerical_features:
    ##Since there are outlier we are going to replace with meidan
    median_values = df[feature].median()
    ##create a new feature to capture NAN values
    df[feature+'nan'] = np.where(df[feature].isnull(), 1, 0) # if nan then keep it as 1, else 0
    df[feature].fillna(median_values, inplace=True)
```
